## Week 3 homework example
This is just an example of how to use the [2024 Living Atlas Dataset](https://github.com/cgiamarino9/PubAff405_S25/blob/main/Weeks/Week03%20Census%20Data/data/LACountyTracts24.zip) from our course website referencing the [data dictionary from Esri](https://www.arcgis.com/home/item.html?id=b0b3b31e531e406185f2de4fff596060). 

**Do not copy this exactly. Use different quantitative columns. Keep ID and geometry columns.**. Be sure to keep `ID` and `geometry` so the tracts can be mapped with Geopandas. 

```python
import geopandas as gpd
```

```python
tracts=gpd.read_file('Data/LACountyTracts24.zip')
```

```python
tracts.head()
```

```python
# define columns to keep; you will have to look at your source's metadata to determine which columns to keep and what they mean
columns_to_keep = ['ID',
                   'TOTPOP_CY',
                   'POPDENS_CY',
                   'DIVINDX_CY',
                   'geometry']

# add it to a new dataframe
tractsCleaned = tracts[columns_to_keep]
```

```python
tractsCleaned.columns = ['FIPS',
               'Total Population (2024)',
               'Population Density (2024',
               'Diversity Index (2024)',
               'geometry']
```

```python
tractsCleaned.head()
```

```python
tractsCleaned.type
```

```python
tractsCleaned.plot()
```

```python
tractsCleaned.plot(figsize=(12,10),
                 column='Diversity Index (2024)',
                 legend=True, 
                 scheme='NaturalBreaks')
```
